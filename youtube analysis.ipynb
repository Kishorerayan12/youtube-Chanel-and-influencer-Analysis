```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df= pd.read_excel('Downloads\youtube analysis.xlsx')
```

```python
df.columns
```

```python
df['Subtitle'].value_counts()
```

```python
plt.figure(figsize= (8,10))
sn.barplot(x='Subtitle', y='No of Likes', data=df)
plt.xlabel('Subtitle', size= 12)
plt.ylabel('Total likes', size=12)
plt.show()
```
